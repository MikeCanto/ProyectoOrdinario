```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CaminataAleatoria(p, n):
    posicion = 0
    probabilidad_paso_adelante = p
    probabilidad_paso_atras = 1 - p
    estados_pisados = [posicion]

    for i in range(1, n):
        step = np.random.choice([-1, 1], p=[probabilidad_paso_atras, probabilidad_paso_adelante])
        posicion += step
        estados_pisados.append(posicion)

    return estados_pisados
```

```python
p = 0.5
n = 100

estados_pisados = CaminataAleatoria(p, n)
```

```python
posiciones = np.array(estados_pisados)
subidas = posiciones[1:] > posiciones[:-1]
bajadas = posiciones[1:] < posiciones[:-1]

plt.figure(figsize=(10, 6))
plt.plot(np.where(subidas)[0], posiciones[1:][subidas], 'g-', label='Subida', linewidth=1)
plt.plot(np.where(bajadas)[0], posiciones[1:][bajadas], 'r-', label='Bajada', linewidth=1)
plt.title("Caminata Aleatoria")
plt.xlabel("Pasos")
plt.ylabel("Posicin")
plt.grid(True)
plt.legend()
plt.show()
```
